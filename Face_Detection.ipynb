```python
import pandas as pd
import numpy as np
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
denis = cv2.imread(r'/content/Denis_Mukwege.jpg')
nadia = cv2.imread(r'/content/Nadia_Murad.jpg')
solvay = cv2.imread(r'/content/solvay_conference.jpg')
```

```python
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
  face_img = img.copy()
  face_rects = face_cascade.detectMultiScale(face_img)# detectMultiScale () function used for detect faces.It takes 3 parameters: Gray: input image (gray scale image) 1.1: scale factor, it specifies how much the image size is reduced with each scale. It improves detection
  for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,0),15)
    
  return face_img
```

```python
result1 = detect_face(denis)
result2 = detect_face(nadia)
result3 = detect_face(solvay)
```

```python
plt.imshow(result1,cmap = 'gray')
```

```python
plt.imshow(result2,cmap = 'gray')
```

```python
plt.imshow(result3,cmap = 'gray')
```

```python
def adj_detect_face(img):
  face_img = img.copy()
  face_rects = face_cascade.detectMultiScale(face_img,scaleFactor =1.1 ,minNeighbors = 5)# detectMultiScale () function used for detect faces.It takes 3 parameters: Gray: input image (gray scale image) 1.1: scale factor, it specifies how much the image size is reduced with each scale. It improves detection
  for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,0),15)
    
  return face_img
```

```python
result4 = adj_detect_face(solvay)
```

```python
plt.imshow(result4)
```

```python
eye_cascade = cv2.CascadeClassifier(r'/content/haarcascade_eye.xml')
```

```python
#function to detect eyes
def detect_eye(img):
  eye_img = img.copy()
  eye_rects = eye_cascade.detectMultiScale(eye_img,scaleFactor =1.1 ,minNeighbors = 5)# detectMultiScale () function used for detect faces.It takes 3 parameters: Gray: input image (gray scale image) 1.1: scale factor, it specifies how much the image size is reduced with each scale. It improves detection
  for (x,y,w,h) in eye_rects:
    cv2.rectangle(eye_img,(x,y),(x+w,y+h),(255,255,0),15)
    
  return eye_img
```

```python
result5 = detect_eye(nadia)
result6 = detect_eye(denis)
```
